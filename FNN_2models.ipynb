```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
# from tensorflow.keras.utils import to_categorical
from sklearn.metrics import classification_report

# Load dataset
data = pd.read_csv("WaterQuality.csv")

# Check for missing values
data = data.dropna()

# Split features and target
X = data.drop("Potability", axis=1)
y = data["Potability"]

# Normalize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split dataset
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
def build_model_a():
    model = Sequential([
        Dense(64, input_dim=X_train.shape[1], activation='relu'),
        Dropout(0.3),
        Dense(32, activation='relu'),
        Dropout(0.3),
        Dense(16, activation='relu'),
        Dropout(0.2),
        Dense(8, activation='relu'),
        Dropout(0.2),
        Dense(1, activation='sigmoid')
    ])
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model

model_a = build_model_a()
history_a = model_a.fit(X_train, y_train, epochs=50, batch_size=32, validation_split=0.2)
```

```python
def build_model_b():
    model = Sequential([
        Dense(128, input_dim=X_train.shape[1], activation='relu'),
        Dropout(0.4),
        Dense(64, activation='relu'),
        Dropout(0.3),
        Dense(32, activation='relu'),
        Dropout(0.2),
        Dense(16, activation='relu'),
        Dropout(0.2),
        Dense(1, activation='sigmoid')
    ])
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model

model_b = build_model_b()
history_b = model_b.fit(X_train, y_train, epochs=50, batch_size=32, validation_split=0.2)
```

```python
from sklearn.metrics import accuracy_score

# Evaluate Model A
print("Model A Performance:")
y_pred_a = (model_a.predict(X_test) > 0.5).astype("int32")
print(classification_report(y_test, y_pred_a))
accuracy_a = accuracy_score(y_test, y_pred_a)
print(f"Model A Accuracy: {accuracy_a:.4f}")

# Evaluate Model B
print("Model B Performance:")
y_pred_b = (model_b.predict(X_test) > 0.5).astype("int32")
print(classification_report(y_test, y_pred_b))
accuracy_b = accuracy_score(y_test, y_pred_b)
print(f"Model B Accuracy: {accuracy_b:.4f}")
```

```python
loss,accuracy=model_a.evaluate(X_test,y_test)
print(f'test accuracy:{ accuracy*100:.2f}%')
```

```python
loss,accuracy=model_b.evaluate(X_test,y_test)
print(f'test accuracy:{ accuracy*100:.2f}%')
```
